```python
from matplotlib import pyplot as plt
```

```python
import json
```

```python
import requests
```

```python
import math
```

```python
tests_file = requests.get('https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-json/dpc-covid19-ita-andamento-nazionale.json').json()
```

```python
tests = []; positive_tests = []; time = []; percentage_positive = []
```

```python
for info in tests_file:
    tests.append(math.log2(info['tamponi']))
    positive_tests.append(math.log2(info['totale_casi']))
    percentage_positive.append((info['totale_casi']/info['tamponi'])*100)
    time.append(info['data'])
```

```python
plt.subplot(3,1,1)
plt.plot(time, tests)
plt.title('tests, positive tests and time')
plt.xlabel('time')
plt.ylabel('tests')


plt.subplot(3, 1, 2)
plt.plot(time, positive_tests)
plt.xlabel('time')
plt.ylabel('positive_tests')

plt.subplot(3, 1, 3)
plt.plot(tests, positive_tests)
plt.xlabel('tests')
plt.ylabel('positive_tests')


plt.savefig('test.png')
plt.show()
```
